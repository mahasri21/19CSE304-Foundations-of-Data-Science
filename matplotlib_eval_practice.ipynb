```python
import numpy as np
x=[0,30,45,60,90]
t=[i*np.pi/180 for i in x]
plt.plot(x,np.sin(t),"+-",label="SINE GRAPH")
plt.plot(x,np.cos(t),"^-",label="COSINE GRAPH")
plt.legend()
plt.show()
```

```python
import numpy as np
x=[0,30,45,60,90]
t=[i*np.pi/180 for i in x]
plt.plot(x,np.sin(t),"+-",label="SINE GRAPH")
plt.plot(x,np.cos(t),"o-",label="COSINE GRAPH")
plt.xticks(x)
plt.legend()
plt.show()
```

```python
import numpy as np
t=np.array([1,2,3,4,5])
plt.plot(t**2,t,color="red")
plt.xlim(0,10)
plt.xlabel("squares")
plt.ylabel("cubes")
plt.title("1st plot")
plt.grid()
plt.xticks([2,4],["A","B","C","D"])
plt.show()
```

```python
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2, t, color="red", label="label1")
plt.plot(t**3, t, color="green", label="label2")
plt.xlim(0,10)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("First plot")
plt.grid()
plt.xticks(np.arange(1,21), ["A", "B", "C", "D", "E"]) #here you can also use array in the ticks area
plt.show()
```

```python
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2, t, color="red", label="label1")
plt.plot(t**3, t, color="green", label="label2")
plt.xlim(0,10)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("First plot")
plt.grid()
plt.tick_params(color="red",pad=20,length=30,axis="both",direction="inout")
plt.minorticks_on()
plt.show()
```
