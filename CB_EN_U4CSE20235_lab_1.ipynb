```python
#q13
owgt = data.loc[(data["Body mass index"] > 25) & (data["Body mass index"] < 30)]
obese = data.loc[data["Body mass index"] > 30]
g1 = owgt.groupby(["Body mass index"])
g2 = obese.groupby(["Body mass index"])


for key, item in g1:
    print(g1.get_group(key), end="\t\t")

for key, item in g2:
    print(g2.get_group(key), end="\t\t")
```
